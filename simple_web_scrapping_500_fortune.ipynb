```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Fortune_500'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
table = soup.find_all('table')[0]
table
```

```python
table_titles = table.find_all('th', scope = 'col')
table_titles
```

```python
table_titles = [title.text.strip() for title in table_titles]
table_titles
```

```python
df = pd.DataFrame(columns = table_titles)
df = df.drop('Rank', axis = 1)
df
```

```python
table_column_data = table.find_all('tr')[1:]
table_column_data
```

```python
for i in table_column_data:
    row_data = i.find_all('td')
    row_data = [row.text.strip() for row in row_data]
    
    length = len(df)
    df.loc[length] = row_data
```

```python
df
```

```python
#df['Rank'] = np.arange(1, len(df) + 1)
df.insert(0, 'Rank', range(1, len(df) + 1))
```

```python
df
```

```python
df.dtypes
```

```python
df['Revenue in USD'] = df['Revenue in USD'].apply(lambda x: x.lstrip('$').rstrip('billion'))
```

```python
df.dtypes
```

```python
df['Revenue in USD'] = df['Revenue in USD'].astype('float')
```
